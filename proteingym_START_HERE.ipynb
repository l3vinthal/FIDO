```python
from env import *
import pandas as pd
import os
```

#### Declare sequence to search against.

```python
metadata = pd.read_csv("/home/ubuntu/cookstore/250603_protein_gym/raw/DMS_ProteinGym_substitutions/DMS_substitutions.csv")
```

#### Run BLAST search.

```python
for study in metadata.iterrows():
    pdb_range = study[1]['pdb_range'].split('-')
    sequence = study[1]['target_seq'][int(pdb_range[0]):int(pdb_range[1])]
    dir = "/home/ubuntu/cookstore/250603_protein_gym/res/" + study[1]['DMS_id'] + "/"
    print ('Running ' + study[1]['DMS_id'])
    print ('Sequence: ' + sequence)
    os.system('python src/blast.py ' + sequence + ' -o ' + dir + 'top5000.fasta -num_aligns 5000')

#!python src/blast.py $sequence -o /home/ubuntu/cookstore/250603_protein_gym/res/YNZC_BACSU_Tsuboyama_2023_2JVD/top5000.fasta -num_aligns 5000
```
